## Conectando con el Arduino

```python
from pyfirmata import Arduino
from time import sleep

board = Arduino('/dev/ttyACM0')

# En la consola otorgar permisos al puerto serial
# ls -la /dev/ttyACM*
# sudo chmod 777 /dev/ttyACM0
```

## Salida Digital

```python
# Para salida digital
# los pines 0 y 1 no son soportados por firmata
# board.digital[pin].write(estado)
# Conectamos un Buzzer a la linea 7
# Alarma acústica ó Parpadeo de un Led

pin=7
tiempo=0.1

try:

    while True:
        board.digital[pin].write(1)
        sleep(tiempo)
        board.digital[pin].write(0)
        sleep(tiempo)
except KeyboardInterrupt:
        board.digital[pin].write(0)
```

```python
# Conectamos cuatro diodos Led a las lineas 3,4,5,6
# board.digital[pin].write(estado)
# Control de motores paso a paso, doble puente en H

tiempo=0.2

def puerto(estado3,estado4,estado5,estado6):
        board.digital[3].write(estado3)
        board.digital[4].write(estado4)
        board.digital[5].write(estado5)
        board.digital[6].write(estado6)
        sleep(tiempo)
try:

    while True:
        puerto(1,0,0,0)
        puerto(0,1,0,0)
        puerto(0,0,1,0)
        puerto(0,0,0,1)

except KeyboardInterrupt:
        puerto(0,0,0,0)
```

```python
from pyfirmata import Arduino
from time import sleep

board = Arduino('/dev/ttyACM0')

# Conectamos un Led a la linea 3
# Variación del brillo de un Led, control de velocidad para motores de CC

pin = board.get_pin('d:3:p') # significa (digital:pin:pwm)
tiempo = 0.2
try:
    while True:
        pwm=0
        for i in range(11):
            pin.write(pwm)
            pwm=pwm+0.1
            sleep(tiempo)
        pwm=1
        for i in range(11,0,-1):
            pin.write(pwm)
            pwm=pwm-0.1
            sleep(tiempo)
except KeyboardInterrupt:
            pin.write(0)
```

```python
from pyfirmata import Arduino
from time import sleep

board = Arduino('/dev/ttyACM0')

# Conectamos un Led RGB a las lineas 9, 10 y 11

pinR = board.get_pin('d:9:p') # significa (digital:pin:pwm)
pinG = board.get_pin('d:10:p')
pinB = board.get_pin('d:11:p') 
tiempo =0.5

def color(red,green,blue):
            pinR.write(red)
            pinG.write(green)
            pinB.write(blue)
            sleep(tiempo)

try:
    while True:
        
            color(0,1,1)       # azul
            color(1,0,1)       # verde
            color(1,1,0)       # rojo

            color(0.75,0.25,0.1) # magenta
            color(1,0,0)
            color(0,1,0)
            color(0,0,1)
            
except KeyboardInterrupt:
            color(1,1,1) # Led apagado
```

## Entrada Analógica

```python
# Conectamos un potenciometro, sensor de temperatura, sensor de luz LDR, sensor de humedad, sensor UV
# en la linea A4

%matplotlib inline
import pylab as pl
from IPython import display
from pyfirmata import Arduino, util
from time import sleep

pl.figure(figsize=(15,5))

board = Arduino('/dev/ttyACM0')

it = util.Iterator(board)
it.start()
board.analog[4].enable_reporting()
analog_4 = board.get_pin('a:4:i') # significa (analogico:pin:input)
t=0
tiempo =0.2

try:
    while True:
        sleep(tiempo)
        dato=analog_4.read()
        pl.plot(t,dato, 'ro')
        t=t+tiempo
        display.clear_output(wait=True)
        display.display(pl.gcf())
        
except KeyboardInterrupt:
        pass
```

## Entrada Digital

```python
# Conectamos un pulsador, un interruptor, un optoacoplador
# en la linea 2

%matplotlib inline
import pylab as pl
from IPython import display
from pyfirmata import Arduino
from time import sleep

pl.figure(figsize=(15,5))

board = Arduino('/dev/ttyACM0')

digital_2 = board.get_pin('d:2:i')
t = 0
tiempo = 0.2

try:
    while True:
        sleep(tiempo)
        dato=digital_2.read()
        pl.plot(t,dato, 'ro')
        t=t+tiempo
        display.clear_output(wait=True)
        display.display(pl.gcf())
        
except  KeyboardInterrupt:
    pass
```
